### [회귀 분석에 사용할 피처 살펴보기]

```python
picher_features_df = picher[['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR',
       '연봉(2018)', '연봉(2017)']]
```

```python
# 피처 각각에 대한 histogram
def plot_hist_each_column(df):
    plt.rcParams['figure.figsize'] = [20, 16]
    fig = plt.figure(1)
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5, 5, i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()
```

# 2. 투수의 연봉 예측하기

### [피처들의 단위 맞춰주기 : 피처 스케일링]

```python
# pandas 형태로 정의된 데이터를 출력할 때, scientific-notation이 아닌 float 모양으로 출력되게 해줍니다.
pd.options.mode.chained_assignment = None
```

```python
# 피처 각각에 대한 scaling을 수행합니다.
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
    return df
```

```python
scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR', '연봉(2017)']
picher_df = standard_scaling(picher, scale_columns)
```

```python
picher_df = picher_df.rename(columns={'연봉(2018)': 'y'})
picher_df.head(5)
```

-----

### [피처들의 단위 맞춰주기 : one-hot-encoding]

```python
team_encoding = pd.get_dummies(picher_df['팀명'])
picher_df = picher_df.drop('팀명',axis = 1)
picher_df = picher_df.join(team_encoding)
```

```python
picher_df.head()
```

### [변수들의 상관관계 분석]

```python
import seaborn as sns

# 변수간의 상관계수 행렬 구하기
corr = picher_df[scale_columns].corr(method = 'pearson')

# corr 행렬 히트맵 시각화
sns.set(font_scale=1.5)
hm = sns.heatmap(corr.values,
            cbar=True,
            annot=True, 
            square=True,
            fmt='.2f',
            annot_kws={'size': 15},
            yticklabels=scale_columns,
            xticklabels=scale_columns)

plt.tight_layout()
plt.show()
```

### [회귀 분석 적용하기]

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train) # 모델 학습
```

# 3. 예측 모델을 평가해보자

### [어떤 피처가 가장 유의미한 변수일까]

```python
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train).fit()

model.summary()
```

```python
mpl.rc('font', family='NanumGothicOTF')
plt.rcParams['figure.figsize'] = [20, 16]

coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

x_labels = model.params.index.tolist()

ax = coefs_series.plot(kind='bar')
ax.set_title('feature_coef_graph')
ax.set_xlabel('x_features')
ax.set_ylabel('coef')
ax.set_xticklabels(x_labels)
```

-----

### [예측 모델의 평가]

```python
X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train) # 모델 학습
```

```python
print(model.score(X_train, y_train)) # train R2 score
print(model.score(X_test, y_test)) # test R2 score
```

```python
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions))) # train RMSE score

y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions))) # test RMSE score
```

### [회귀분석 예측 성능을 높이기 위한 방법 : 다중공선성 확인]

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```

```python
picher = pd.read_csv(picher_file_path)
picher = picher[['선수명', '연봉(2017)']]

tf = picher_df.sort_values(by=['y'], ascending=False)
tf.drop(['연봉(2017)'], axis = 1, inplace = True, errors = 'ignore')
tf = tf.merge(picher, on=['선수명'], how='left')
tf = tf[['선수명', 'y', '예측연봉(2018)', '연봉(2017)']]
tf.columns = ['선수명', '실제연봉(2018)', '예측연봉(2018)', '작년연봉(2017)']

# 재계약하여 연봉이 변화한 선수만을 대상으로 관찰
tf = tf[tf['작년연봉(2017)'] != tf['실제연봉(2018)']]
tf = tf.reset_index()
tf = tf.ix[:10, :]
tf.head(10)
```

```python
tf.plot(x='선수명', y=['작년연봉(2017)', '예측연봉(2018)', '실제연봉(2018)'], kind="bar")
```
